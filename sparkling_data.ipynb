```python
#Import modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the csv file downloaded from https://wineinstitute.org/our-industry/statistics/us-sparkling-wine-champagne-shipments/

sparkling_data = pd.read_csv('sparkling_wine_data.csv')
sparkling_data
```

```python
sparkling_data.set_index(['Year'])
```

```python
# Clean the "California Winery Sparkling" column
sparkling_data.dtypes
```

```python
# Change datatype of Year
sparkling_data['California Winery Sparkling'] = sparkling_data['California Winery Sparkling'].str.replace(',', '').astype(int)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sparkling_data['Year'], sparkling_data['California Winery Sparkling'], marker='o', linestyle='-', color='yellow')
plt.title('California Winery Sparkling Production Over the Years')
plt.xlabel('Year')
plt.ylabel('California Winery Sparkling')
plt.grid(True)

# Set the background color to gray
plt.gca().set_facecolor('gray')

#save the fig
plt.savefig("Cali_Spklg_Production.png")

#show the plot
plt.show()
```
